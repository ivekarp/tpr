Вы работаете секретарем и вам постоянно приходят различные документы. Вы должны быть очень внимательны, чтобы не потерять ни один документ. Каталог документов хранится в следующем виде:

```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}

def people_document():
    number = input('Введите номер документа: ')
    for doc in documents:
        if (doc['number'] == number):
            print('Владелец документа под номером ' + number + ' : ' + doc['name'])
            break
    else:
        print('Такого документа нет!')


def shelf():
    number = input('Введите номер документа: ')
    found = False
    for shelf_directorie in directories.items():
        for doc_number in shelf_directorie[1]:
            if (doc_number == number):
                print('Документ ' + number + ' лежит на полке ' +  shelf_directorie[0])
                found = True
                break
        if (found == True):
            break
    else:
        print('Такого документа нет!')

        
def l_list():
    for person in documents:
        print(person['type'] + ' ' + person['number'] + ' ' + person['name'])
    

def add_shelf():
    directories_number = input('Введите номер новой полки: ')
    if (directories_number not in directories):
        directories[directories_number] = []
        print('Полка номер ' + directories_number + ' создана!')
    else:
        print('Такая полка уже существует!')


def delete():
    initial_len = len(documents)
    doc_number = input('Введите номер документа: ')
    for i, d in enumerate(documents):
        if d['number'] == doc_number:
            documents.pop(i)
    if (initial_len == len(documents)):
        print('Документ не существует')
    for key, value in directories.items():
        if (doc_number in value):
            value.remove(doc_number)
            print('Документ успешно удален')


def move():
    doc_number = input('Введите номер документа: ')
    shelf_number = input('Введите номер целевой полки: ')
    doc_existence = False
    if (shelf_number not in directories):
        print('Полки не существует')
    for key, value in directories.items():
        if (doc_number in value):
            doc_existence = True
            directories[shelf_number] += [doc_number]
            value.remove(doc_number)


def add():
    doc_type = input('Введите тип документа: ')
    doc_number = input('Введите номер документа: ')
    doc_owner = input('Введите имя: ')
    shelf_number = input('Введите номер полки: ')
    if (shelf_number not in directories):
        print('Полки не существует')
    new_doc = dict(type=doc_type, number=doc_number, name=doc_owner)
    documents.append(new_doc)
    directories[shelf_number] += [doc_number]
    print('Документ успешно добавлен')


def print_documents(documents):
    for doc in documents:
        print(doc)

        
def print_directories(directories):
    for key,value in directories.items():
        print(key,value)

        
def help_command():
    info = '''Помощь по командам:
        p) people) - Вывести имя человека, которому принадлежит документ
        s) shelf) - Вывести номер полки, на которой находится документ
        l) list) - Вывести список всех документов
        as) add_shelf) - Добавить новую полку в перечень
        d) delete) - Удалить документ из каталога и перечня полок
        m) move) - Переместить документ с текущей полки на целевую
        add) - Добавить новый документ в каталог и перечень полок
        e) exit) - Выход
    '''
    print(info)

    
def menu():
    help_command()
    command = input('Введите команду: ')
    if (command == 'p') or (command == 'people'):
        people_document()
        print()
    elif (command == 's') or (command == 'shelf'):
        shelf()
        print()
    elif (command == 'l') or (command == 'list'):
        l_list()
        print()
    elif (command == 'as') or (command == 'add_shelf'):
        add_shelf()
        print()
    elif (command == 'd') or (command == 'delete'):
        delete()
        print()
    elif (command == 'm') or (command == 'move'):
        move()
        print()
    elif (command == 'add'):
        add()
        print()
    elif (command == 'e') or (command == 'exit'):
        return 0
    elif (command == '3'):
        print(directories)
    else:
        print('Ошибка! Введите правильную команду!\n')
    
    
def main():
    while (True):
        if (menu() == 0):
            break
        
main()
```
