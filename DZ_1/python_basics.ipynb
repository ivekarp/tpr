## Задание 2

Дана переменная, в которой хранится четырехзначное число (год). Необходимо написать программу, которая выведет, является ли данный год високосным или обычным.

Пример работы программы:
1.  

```
year = 2020
```
Результат:  
`Високосный год` 

2.  

```
year = 2019
```
Результат:  
`Обычный год` 

````python
# Интересное задание
# Тут надо использовать алгоритм определения високосного года
# Ежели:
# 1) Год не делится нацело на 4
# 2) Делится нацело на 100 и не делится нацело на 400
# Тогда: год - обычный
# В противном случае: год - високосный

year = int(input('Введите год:'))
if (year % 4 != 0) or (year % 100 == 0 and year % 400 != 0):
    print(str(year) + " - это обычный год")

else:
    print(str(year) + " - это високосный год")
````

## Задание 5

Дана переменная, в которой хранится шестизначное число (номер проездного билета). Напишите программу, которая будет определять, является ли данный билет "счастливым". Билет считается счастливым, если сумма первых трех цифр совпадает с суммой последних трех цифр номера.   

Примеры работы программы:  
1. ```number = 123456```

Результат:  
`Неасчастливый билет`
2. ```number = 123321```

Результат:  
`Счастливый билет`


````python
# Тут будет удобнее написать функцию
# Всё просто: берём число, складываем остатки от деления на 10
# И целочисленно делим на 10 само число
# Возвращаем сравнение сумм


def isLucky(number):
    s1=0
    s2=0
    for i in range(3):
        s1 += number % 10
        number //= 10
    for i in range(3):
        s2 += number % 10
        number //= 10
    return (s1==s2)
 
n = int(input('Введите номер проездного билета'))

if (isLucky(n)):
    print(str(n)+" это счастливый билет")
else:
    print(str(n)+" это обычный билет")
````

## Задание 6

Напишите программу, которая сможет вычислять площади трех фигур (круг, треугольник и прямоугольник).
Тип фигуры запрашиваем через пользовательский ввод, после чего делаем запрос характеристик фигуры:

- если пользователь выбрал круг, запрашиваем его радиус,
- если треугольник – длины трех его сторон;
- если прямоугольник – длины двух его сторон.

Пример работы программы:
1.  

```
Введите тип фигуры:
Круг

Введите радиус круга:
10
```
Результат:  
`Площадь круга: 314.16`

2.  

```
Введите тип фигуры:
Треугольник

Введите длину стороны A:
2

Введите длину стороны B:
2

Введите длину стороны C:
3
```
Результат:  
`Площадь треугольника: 1.98`

````python
# Тут уже будет удобнее написать несколько функций
# И само собой подключить математику для числа Пи и корня


import math

def figure(name):
    if (name == 'Круг'):
        circle()
    elif (name == 'Треугольник'):
        triangle()
    elif (name == 'Прямоугольник'):
        rectangle()
    else:
        print('Ошибка!')

def circle():
    radius = float(input('Введите радиус круга:'))
    area = math.pi * radius ** 2
    print('Площадь круга = ' + str(area))

def triangle():
    a = float(input('Введите длину стороны A:'))
    b = float(input('Введите длину стороны B:'))
    c = float(input('Введите длину стороны C:'))
    p = 0.5 * (a + b + c) # Полупериметр для формулы Герона
    area = math.sqrt(p * (p - a) * (p - b) * (p - c)) # Формула Герона для нахождения площади произвольного треугольника
    print('Площадь треугольника = ' + str(area))

def rectangle():
    a = float(input('Введите длину стороны A:'))
    b = float(input('Введите длину стороны B:'))
    area = a * b
    print('Площадь прямоугольника = ' + str(area))

name = input('Введите тип фигуры:')
figure(name)
````
